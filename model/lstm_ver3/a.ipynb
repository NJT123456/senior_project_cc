```python
import pandas as pd
import numpy as np
```

# train 80

```python
file_names = ['data_lstm_mse_lag5.csv', 'data_lstm_mse_lag10.csv', 'data_lstm_mse_lag15.csv', 'data_lstm_mse_lag20.csv', 'data_lstm_mse_lag25.csv', 'data_lstm_mse_lag30.csv']
dfs = []

for file_name in file_names:
    df = pd.read_csv(file_name)
    dfs.append(df)

data = pd.concat(dfs)
data = data.fillna('')
data
```

```python
data = data.drop(columns=["Drop out2","Third layer"])
data
```

## MAE

```python
mae_lstm = data.groupby(['Lags'])['MAE error'].transform(min) == data['MAE error']
data[mae_lstm]
```

```python
data80 = data[mae_lstm]
data80.to_csv('data_error.csv',index=False)
```

```python
data1 = data[mae_lstm]
data1 = data1['MAE error'].mean()
print('Average MAE : %.5f' %data1)
```

```python
datamae = data[mae_lstm]
datamae = datamae.drop(columns=["MAPE error","MSE error","RMSE error"])
datamae
```
